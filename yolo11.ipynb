```python
import os
import cv2
import matplotlib.pyplot as plt

#  path  save_dir 
save_dir = '/Users/oattao/Desktop/runs/detect/predict13'

# 
if os.path.exists(save_dir):
    # 
    files = os.listdir(save_dir)
    print(f"Files in {save_dir}: {files}")

    #  ( .jpg  .png)
    image_files = [file for file in files if file.endswith(('.jpg', '.png', '.jpeg'))]

    #  
    if image_files:
        image_path = os.path.join(save_dir, image_files[0])  # 
        image = cv2.imread(image_path)

        #  BGR  RGB 
        image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

        #  Matplotlib
        plt.figure(figsize=(10, 8))
        plt.imshow(image_rgb)
        plt.axis('off')  # 
        plt.title(f"Annotated Image: {image_files[0]}")
        plt.show()
    else:
        print(f"No image files found in {save_dir}")
else:
    print(f"Directory {save_dir} does not exist.")
```
