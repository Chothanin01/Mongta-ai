```python
import os
import cv2
import matplotlib.pyplot as plt

# กำหนด path ของ save_dir ที่บันทึกไฟล์ภาพ
save_dir = '/Users/oattao/Desktop/runs/detect/predict13'

# ตรวจสอบว่าโฟลเดอร์นี้มีไฟล์หรือไม่
if os.path.exists(save_dir):
    # แสดงรายชื่อไฟล์ในโฟลเดอร์
    files = os.listdir(save_dir)
    print(f"Files in {save_dir}: {files}")

    # ตรวจสอบว่าโฟลเดอร์นี้มีไฟล์ภาพไหม (สมมติว่าเป็น .jpg หรือ .png)
    image_files = [file for file in files if file.endswith(('.jpg', '.png', '.jpeg'))]

    # หากมีไฟล์ภาพในโฟลเดอร์ ให้แสดงภาพนั้น
    if image_files:
        image_path = os.path.join(save_dir, image_files[0])  # ใช้ไฟล์แรกในรายการ
        image = cv2.imread(image_path)

        # แปลงจาก BGR เป็น RGB สำหรับการแสดงผล
        image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

        # แสดงภาพใน Matplotlib
        plt.figure(figsize=(10, 8))
        plt.imshow(image_rgb)
        plt.axis('off')  # ปิดแสดงแกน
        plt.title(f"Annotated Image: {image_files[0]}")
        plt.show()
    else:
        print(f"No image files found in {save_dir}")
else:
    print(f"Directory {save_dir} does not exist.")
```
